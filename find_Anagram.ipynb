```python
words=[]
for line in open ('words.txt','r'):
  words.append(line.lower().strip())
```

```python
# In order to avoid duplicates will get set instead of List 
words= set()
for line in open('words.txt','r'):
  words.add(line.strip().lower())
  
```

```python
#  writing it in comprehensive way
words= sorted({line.strip().lower() for line in open('words.txt','r')})
```

```python
words[:-10]
```

```python
def signature(word):
  return "".join(sorted(word))
```

```python
def find_anagram(myword):
  
  my_sig=signature(myword)

  for word in words:
    if my_sig == signature(word):
      print(word)
   
```

```python
find_anagram('doar')
```

```python
%time find_anagram('dictionary')
```

```python
import collections
# In order to reduce performance time we will create dictionary of words indexing it by signature of words. 

sig_dict= collections.defaultdict(set)

for word in words:
  sig_dict[signature(word)].add(word)
```

```python
# Removing single words from dictionary because they will be nagram of themseleves so no need to search through them
anagram_by_signature={sig : sorted(wordsheet,key=len) for sig, wordsheet in sig_dict.items() if len(wordsheet)>1}
```

```python
def find_anagram_fast(myword):
  sig= signature(myword)
  return anagram_by_signature[sig]
```

```python
# Much faster performance compared to earlier version
%time find_anagram_fast('dictionary')
```

```python
# to resolve the key error for word not present in the dict will add try except block

def find_anagram_fast(myword):
  try:
    sig= signature(myword)
    return anagram_by_signature[sig]
  except KeyError as ke: 
    return set()
```
